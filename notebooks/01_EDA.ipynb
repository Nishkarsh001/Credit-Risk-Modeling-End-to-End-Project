```python
import pandas as pd
import numpy as np

# Load your existing data
df = pd.read_csv('../data/raw/german_credit_data.csv')
df = df.drop('Unnamed: 0', axis=1)

# Create Risk column based on financial rules
def assign_risk(row):
    risk_score = 0
    
    # Credit amount risk
    if row['Credit amount'] > 10000:
        risk_score += 3
    elif row['Credit amount'] > 5000:
        risk_score += 2
    elif row['Credit amount'] > 2000:
        risk_score += 1
        
    # Duration risk
    if row['Duration'] > 36:
        risk_score += 3
    elif row['Duration'] > 24:
        risk_score += 2
    elif row['Duration'] > 12:
        risk_score += 1
    
    # Age risk
    if row['Age'] < 25:
        risk_score += 2
    elif row['Age'] < 30:
        risk_score += 1
        
    # Saving accounts risk
    if pd.isna(row['Saving accounts']) or row['Saving accounts'] == 'little':
        risk_score += 2
    elif row['Saving accounts'] == 'moderate':
        risk_score += 1
        
    # Checking account risk
    if pd.isna(row['Checking account']) or row['Checking account'] == 'little':
        risk_score += 2
    elif row['Checking account'] == 'moderate':
        risk_score += 1
    
    # Assign risk
    if risk_score >= 6:
        return 'bad'
    else:
        return 'good'

df['Risk'] = df.apply(assign_risk, axis=1)

# Fix --- use .loc['good'] not [0]
good_count = df['Risk'].value_counts()['good']
bad_count = df['Risk'].value_counts()['bad']

print("Risk Distribution:")
print(df['Risk'].value_counts())
print(f"\nImbalance Ratio: {good_count/bad_count:.2f}:1")

# Save updated dataset
df.to_csv('../data/raw/german_credit_data.csv', index=False)
print("\nSaved with Risk column!")
df.head()
```
